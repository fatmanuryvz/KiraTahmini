```python
def drop_columns(df, cols):
    """
    This function removes the columns from a dataFrame
    """
    for col in cols :
        try:
            df.drop([col], axis=1, inplace=True)
        except Exception as e:
            print(f"bir hata oluştu:{e}")
```
